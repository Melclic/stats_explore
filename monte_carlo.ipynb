```python
import statsmodels
import numpy as np
import pandas as pd
import random
import plotly.express as px
import scipy
```

```python
mice_pheno = pd.read_csv('flatfiles/mice_pheno.csv')
```

```python
m = mice_pheno[mice_pheno['Sex']=='F']
m = m[m['Diet']=='chow']
bw = m['Bodyweight'].to_list()
```

```python
np.var(random.sample(bw, 10))
```

```python
def ttestgenerator(l, n):
    cases = random.sample(l, n)
    controls = random.sample(l, n)
    tstat = (np.mean(cases)-np.mean(controls))/np.sqrt(np.var(controls)/n)
    return tstat
```

```python
ttests = [ttestgenerator(bw, 100) for i in range(1000)]
```

```python
fig = px.histogram(x=ttests)
fig.show()
```

#here we want to make a parametric simualtion... ie. assume that the population behaves in some way and simulate the data

```python
def ttestgenerator_param(n, mean, sd):
    cases = np.random.normal(mean, sd, n)
    controls = np.random.normal(mean, sd, n)
    tstat = (np.mean(cases)-np.mean(controls))/np.sqrt(np.var(cases)/n + np.var(controls)/n)
    return tstat
```

```python
ttests = [ttestgenerator_param(100, 24, 3.5) for i in range(1000)]
```

```python
fig = px.histogram(x=ttests)
fig.show()
```
